### Parte 2:
- A quantidade de excees que existem em Python  imensa. Basta voc ir, agora mesmo no Google, e pesquisar que voc ver. (https://docs.python.org/3/library/exceptions.html)
- Mas no se preocupe! Voc no tem que saber todas as excees. Basta lembrar que toda a exceo, no caso do Python,  filho de uma classe maior chamada **Exception**.
- Afim de corrigir essas excees, vamos aprender como **tentar** antes de **fazer**. Em outras palavras, ao invs de executar linhas de comando imperativamente, vamos usar o mdulo **try** assim como na figura a seguir.


- Em resumo, o bloco onde temos o **try** nos remete a fazer o compilador tentar fazer alguma coisa. Porm, caso no consiga, verificamos o segundo bloco, dentro de **except**. E caso d certo, vamos para o **else**. J em **finally**  o que vai acontecer independente se deu certo ou se deu erro. 
-  importante lembrar que as clusulas **else** e **finally** so opcionais. Nem sempre  preciso us-las.
- Vejamos um exemplo a seguir:

```python
# EXEMPLO 1
try:
    numerador = int(input("Numerador: "))
    denominador = int(input("Denominador: "))
    resposta = numerador/denominador
except:
    print("\033[31mErro\033[m: Tivemos um 'bug' no sistema.")
else:
    print(f"O resultado  {resposta:.2f}")
finally:
    print(f"{'< FIM DO PROGRAMA >':=^40}")
```

### Parte 3:
- Inclusive, ns podemos tratar o erro que estamos enfretando, como veremos no prximo exemplo:

```python
# EXEMPLO 2
try:
    numerador = int(input("Numerador: "))
    denominador = int(input("Denominador: "))
    resposta = numerador/denominador
except Exception as erro:
    print(f"O problema encontrado foi: {erro.__class__}")
else:
    print(f"O resultado  {resposta:.2f}")
finally:
    print(f"{'< FIM DO PROGRAMA >':=^40}")
```

### Parte 4:
- Para cada erro, ns podemos criar blocos de "Exception" diferente, tratando cada erro respectivamente. E mesmo em um **Exception** podemos, separando-os por vrgula, colocar erros especficos que gostariamos de tratar.


- Agora vamos treinar isso no exemplo que estamos criando:

```python
# EXEMPLO 3
try:
    numerador = int(input("Numerador: "))
    denominador = int(input("Denominador: "))
    resposta = numerador/denominador
    
except (ValueError, TypeError):
    print("Tivemos um problema com os tipos de dados que voc digitou.")
    
except ZeroDivisionError:
    print("No  possvel diviso por zero.")
    
except KeyboardInterrupt:
    print("O usurio preferiu no informar os dados.")
    
else:
    print(f"O resultado  {resposta:.2f}")
    
finally:
    print(f"{'< FIM DO PROGRAMA >':=^40}")
```

```python
# EXEMPLO 4

while True:
    try:
        numero = int(input("Digite um valor: "))
    except (ValueError, TypeError):
        print("Erro: Tente novamente.")
        continue
    else: 
        print(f"O valor digitado foi {numero}.")
        break
```

```python
# EXEMPLO 5

def inputNumber(message):
    while True:
        try:
            userInput = int(input(message))
        except ValueError:
            print("No  um nmero inteiro! TENTE NOVAMENTE!")
            continue
        else:
            return userInput
            break

numero = inputNumber("Digite um nmero inteiro: ")
print(f'Voc finalmente digitou um valor inteiro e igual a {numero}.')
```
