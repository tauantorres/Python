## Módulos e Pacotes
- Basicamente, com a evolução da programação, os programas foram ficando cada vez maiores e maiores. Sendo assim, a criação dos módulos era algo mais do que natural nesse processo evolutivo. Visto que, o foco dessa criação era:
1. Dividir programas grandes;
2. Aumentar a legibilidade;
3. Facilitar a manutenção.

## Parte 1:
- Vá até o seu Pycharm e digite o seguinte código:

```python
def fatorial(numero):
    produto = 1
    for i in range(1, numero + 1):
        produto *= i
    return produto

# MAIN FUNCTION:
numero = int(input("Digite um valor: "))
print(f"O fatorial de {numero} é {fatorial(numero)}.")
```

## Parte 4:
- Agora nós podemos falar das vantagens da modularização:
1. Organização do código;
2. Facilidade na manutenção;
3. Ocultação de código detalhado;
4. Reutilização em outros projetos;

- Porém, quando só os **módulos** não puderem mais satisfazer as necessidade do programador, nós temos a solução, dentro do Python, que nós chamamos de **Pacotes**(aquilo que em muitas linguagens são chamados de ***Bibliotecas***). Sendo assim, vamos ao seguinte exemplo prático para que você possa entender o que são **Pacotes**:

> - Imagine que você já criou um módulo, porém ele está enorme e com muitas funções. Porém, visto que o objetivo da modularização é diminuir o código principal e guardar dentro de um módulo. Sendo assim, se ele for muito grande, voltam todos os problemas anteriores.


>- Uma solução que se encontrava para contornar esse problema, era criar mais módulos afim de diminuir a sobrecarga de um único módulo só. O Python criou uma forma de juntar os módulos e separá-los por assuntos. Sim, ele mesmo, o **Pacote**.


>- Observe que nós poderiamos escrever os seguites comandos:
>> 1. import uteis;
>> 2. from uteis import datas;
>- O primeiro me importaria todos os pacotes de uteis. Já o segundo, apenas as funcionalidade que cabem ao grupo datas.
